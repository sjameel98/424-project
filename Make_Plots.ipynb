```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dateutil import parser

from datetime import datetime

sns.set_theme()
```

```python
def get_time(df, num_epochs):
    # F
    dt = parser.parse(list(df['time'])[-1]) - parser.parse(list(df['time'])[0])
    secs = dt.total_seconds()
    return secs/num_epochs

def get_acc(df):
    # Find best test accuracy
    return max(list(df['test acc']))

def plot_loss(dfs, labels, title, filename):
    new_dfs = []
    for df in dfs:
        new_dfs.append(df.iloc[:20])
    plt.figure(figsize=(10, 5))
    for i in range(len(dfs)):
        initial_entry = pd.DataFrame({'test loss': [3.14], 'train loss': [3.14]})
        df1_test = pd.concat([initial_entry, new_dfs[i]]).reset_index()
        plt.plot(df1_test['test loss'], label=labels[i])
    plt.title(title)
    plt.ylabel('Test Loss')
    plt.xlabel('Epoch')
    plt.legend()
    plt.ylim((0, 2.5))
    plt.xlim((-1, 21))
    plt.savefig(filename)

def plot(dfs, labels, title, filename):
    #plt.clf()
    new_dfs = []
    for df in dfs:
        new_dfs.append(df.iloc[:20])
    plt.figure(figsize=(10, 5))
    for i in range(len(dfs)):
        initial_entry = pd.DataFrame({'test acc': [10], 'train acc': [10]})
        df1_test = pd.concat([initial_entry, new_dfs[i]]).reset_index()
        plt.plot(df1_test['test acc'], label=labels[i])
    plt.title(title)
    plt.ylabel('Test Accuracy (%)')
    plt.xlabel('Epoch')
    plt.legend()
    plt.ylim((40, 100))
    plt.xlim((-1, 21))
    plt.savefig(filename)
```

```python
### GoogleNet CIFAR10
# Run 21 -- sgd cifar10
df1 = pd.read_json("mie424_results_v5/values_run-21.json")

# Run 22 -- sgd w/ lookahead cifar10
df2 = pd.read_json("mie424_results_v5/values_run-22.json")

# Run 27 -- adam cifar10
df3 = pd.read_json("mie424_results_v5/values_run-27.json")

# Run 28 -- adam w/ lookahead cifar10
df4 = pd.read_json("mie424_results_v5/values_run-28.json")
print(df4.shape)

# Run 31 -- apollo (milestone) cifar10 
df5 = pd.read_json("mie424_results_v5/values_run-31.json")
print(df5.shape)

# Run 32 -- apollo (milestone) w/ lookahead cifar10
df6 = pd.read_json("mie424_results_v5/values_run-32.json")
print(df6.shape)

# Run 39 -- apollo (cosine) cifar10 
df7 = pd.read_json("mie424_results_v5/values_run-39.json")
print(df7.shape)

# Run 40 -- apollo (cosine) w/ lookahead cifar10
df8 = pd.read_json("mie424_results_v5/values_run-40.json")
print(df8.shape)


print(get_time(df1, 50))
print(get_time(df2, 50))
print(get_time(df3, 50))
print(get_time(df4, 50))
print(get_time(df5, 20))
print(get_time(df6, 20))
print(get_time(df7, 20))
print(get_time(df8, 20))


print('\n\n\n')
print(get_acc(df1))
print(get_acc(df2))
print(get_acc(df3))
print(get_acc(df4))
print(get_acc(df5))
print(get_acc(df6))
print(get_acc(df7))
print(get_acc(df8))

plot([df1, df2, df3, df4, df5, df6, df7, df8], 
     ['SGD', 'SGD w/ Lookahead', 'Adam', 'Adam w/ Lookahead', 
      'Apollo (milestone)', 'Apollo (milestone) w/ Lookahead',
      'Apollo (cosine)', 'Apollo (cosine) w/ Lookahead'], 'GoogLeNet on CIFAR-10 Test Accuracy', 'googlenet_cifar_acc_v2.png')

plot_loss([df1, df2, df3, df4, df5, df6, df7, df8], 
     ['SGD', 'SGD w/ Lookahead', 'Adam', 'Adam w/ Lookahead', 
      'Apollo (milestone)', 'Apollo (milestone) w/ Lookahead',
      'Apollo (cosine)', 'Apollo (cosine) w/ Lookahead'], 'GoogLeNet on CIFAR-10 Test Loss', 'googlenet_cifar_loss_v2.png')







```
